```python
import pandas as pd
import numpy as np
import glob
import pandas_profiling
```

```python
files = glob.glob('data/formatted_labs/*')
files.remove('data/formatted_labs/formatted_patients_palpation.csv')
files
```

```python
data = [pd.read_csv(file, index_col=0) for file in files]
```

```python
for df in data:
    print(df.describe(include='all'))
```

```python
patients = pd.read_excel('data/background_info.xlsx')
pd.to_datetime(patients.BIRTH_DATE)
```

```python
patients.profile_report()
```

```python
features = []
for pt in patients['PT_ID']:
    pt_data = [df[df['PT_ID'] == pt] for df in data]
    feature = []
    for pt_data_df in pt_data:
        recent_data = pt_data_df[pt_data_df['TIME_TAKEN'] == pt_data_df['TIME_TAKEN'].max()].T.squeeze()
        if recent_data.empty == False:    
            feature.append(recent_data.values[3])
        else:
            feature.append(-1)
    print(feature)
    features.append(feature)
```
